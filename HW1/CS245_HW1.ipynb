```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import pandas as pd
```

```python
min_sup = 3
item_list = ["Python", "Machine Learning", "Introduction to AI", "JavaScript", "C++", "Data Science", "Mathematics"]
t1 = ["Python", "Machine Learning"]
t2 = ["Introduction to AI", "JavaScript", "C++"]
t3 = ["Python", "Machine Learning", "Data Science", "Introduction to AI", "JavaScript"]
t4 = ["Mathematics", "Machine Learning"]
t5 = ["Python", "Data Science", "Introduction to AI", "Machine Learning"]
t6 = ["Python", "Data Science"]
t7 = ["Mathematics", "Machine Learning", "Data Science"]
t8 = ["Python", "Mathematics"]
t9 = ["Python", "Machine Learning", "Introduction to AI"]
t10 = ["Introduction to AI", "Mathematics", "C++"]
```

```python
T_list = [["Python", "Machine Learning"], 
          ["Introduction to AI", "JavaScript", "C++"], 
          ["Python", "Machine Learning", "Data Science", "Introduction to AI", "JavaScript"], 
          ["Mathematics", "Machine Learning"], 
          ["Python", "Data Science", "Introduction to AI", "Machine Learning"], 
          ["Python", "Data Science"], 
          ["Mathematics", "Machine Learning", "Data Science"], 
          ["Python", "Mathematics"], 
          ["Python", "Machine Learning", "Introduction to AI"], 
          ["Introduction to AI", "Mathematics", "C++"]]
```

```python
te = TransactionEncoder()
te_transform = te.fit(T_list).transform(T_list)
T_df = pd.DataFrame(te_transform, columns = te.columns_)
```

```python
T_df
```

```python
freq_itemsets = apriori(T_df, min_support = 0.3, use_colnames = True)
freq_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(freq_itemsets, metric="confidence", min_threshold=0.6)
```
